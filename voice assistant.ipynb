```python
import speech_recognition as sr
import nltk
from nltk.corpus import stopwords
import pyttsx3

def initialize_nltk():
    # Download NLTK resources
    nltk.download('stopwords')

def get_audio():
    recognizer = sr.Recognizer()

    with sr.Microphone() as source:
        print("Listening...")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)

    return audio

def recognize_speech(audio):
    recognizer = sr.Recognizer()

    try:
        text = recognizer.recognize_google(audio)
        print(f"You said: {text}")
        return text
    except sr.UnknownValueError:
        print("Sorry, I could not understand what you said.")
        return ""
    except sr.RequestError as e:
        print(f"Error connecting to Google API: {e}")
        return ""

def process_command(command):
    stop_words = set(stopwords.words('english'))

    if any(word in command.lower() for word in ['hello sankalp', 'hello sankalp']):
        return "Hello shu ! How can I help you today?"
    elif any(word in command.lower() for word in ['how r u sankalp', 'how are you sankalp']):
        return " i am fine shu ! i hope you also good"
    elif any(word in command.lower() for word in ['thanks sankalp', 'thanks sankalp']):
        return "You're welcome shu!"
    elif any(word in command.lower() for word in ['bye', 'goodbye']):
        return "Goodbye! Have a great day."
    else:
        return "I'm sorry, I didn't understand that."

def speak(text):
    engine = pyttsx3.init()
    engine.say(text)
    engine.runAndWait()

if __name__ == "__main__":
    initialize_nltk()

    while True:
        audio_input = get_audio()
        command = recognize_speech(audio_input)

        if command:
            response = process_command(command)
            print("Assistant:", response)
            speak(response)
```
